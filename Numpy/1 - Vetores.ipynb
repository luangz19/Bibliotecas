```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
u = np.array([1,2])
v = np.array([3,4])
```

```python
u
```

```python
v
```

```python
# soma de vetores
s = u + v
s
```

```python
# diferença de vetores
d = u - v
s
```

```python
# mutiplicação de um escalar por um vetor
a = 2
a*v
```

```python
b = 3
b*u
```

```python
a*v + b*u
```

```python
plt.style.use("dark_background")
for i in [u,v,s,d]:
    cord1, cord2 = (i[0], i[1])
    plt.quiver(0,0,cord1,cord2,color="yellow", scale_units="xy", angles="xy", scale=1)
    plt.scatter(cord1,cord2,  label=f"vetor = {i}")
plt.xlim(-4,8)
plt.ylim(-4,8)
plt.legend()
```

```python
x = np.array([1,0,0])
y = np.array([0,1,0])
z = np.array([0,0,1])
```

```python
ax = plt.axes(projection="3d")
for i in [x,y,z]:
    cord1, cord2, cord3 = (i[0], i[1],i[2])
    ax.quiver(0,0,0,cord1,cord2,cord3, color="yellow", arrow_length_ratio=0.02)
    ax.scatter(cord1,cord2,  label=f"vetor = {i}")
ax.set_xlim(1,0)
ax.set_ylim(0,1)
ax.set_zlim(0,1)
plt.legend()
```
