```python
import matplotlib.pyplot as plt
from numpy import random
import seaborn as sns
import pandas as pd
```

```python
precos = random.randint(25000,100000,(10,10))
```

```python
bairros = ["Bairro-" + str(x+1) for x in range(10)]
```

```python
casas = ["Valor" for Valor in range(10)]
```

```python
df = pd.DataFrame(precos, index=casas, columns=bairros)
df
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(7,5))
sns.boxplot(df)
plt.tight_layout()
```

```python
sns.boxplot(df, orient="h")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.boxplot(df)
plt.subplot(2,2,2)
sns.boxplot(df, orient="y")
plt.tight_layout()
```
