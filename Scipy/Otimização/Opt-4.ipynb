```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
```

```python
inicio = 0
final = 2*np.pi
incremento = 0.5
x = np.arange(inicio,final,incremento)
a = 2
b = 10
np.random.seed()
y_noise = 0.2 * np.random.normal(size=x.size)
y = a * np.sin(x + b)
y = y + y_noise
plt.style.use("dark_background")
plt.plot(x,y, 'or')
```

```python
def Model(t,a,b):
    y = a*np.cos(t) + b*np.sin(t)
    return y
```

```python
popt,pcov = curve_fit(Model,x,y)
```

```python
print(popt)
print(pcov)
```

```python
y_pred = Model(x,*popt)
RMSE = root_mean_squared_error(y_pred,y)
MAE = mean_absolute_error(y_pred,y)
R2 = r2_score(y_pred,y)
print(f"RMSE = {RMSE}")
print(f"MAE = {MAE}")
print(f"R² = {R2}")
```

```python
x_data = np.arange(inicio,final,0.1)
y_data = Model(x_data,*popt)
plt.plot(x,y,"o")
plt.plot(x_data,y_data)
```
