```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import scipy.optimize as opt
import numpy as np
# %matplotlib qt
```

```python
# Minimizar f(x,y) = (a -x)**2 + b*(y - x**2)**2
a,b = (1,100)
f = lambda x,y: (a -x)**2 + b*(y - x**2)**2
```

```python
x,y = (np.linspace(-2,2,100),np.linspace(-2,2,100))
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
plt.style.use("dark_background")
ax = plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="Oranges")
```

```python
# Minimizar f(x,y) = (a -x)**2 + b*(y - x**2)**2 onde x[0] = x e x[1] = y
a,b = (1,100)
f = lambda x: (a -x[0])**2 + b*(x[1] - x[0]**2)**2
```

```python
x0 = [-1.2,1]

xopt = opt.fmin(f,x0)
```

```python
print(xopt)
```

```python
plt.style.use("dark_background")
ax = plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="Oranges",alpha=0.5)
ax.scatter(xopt[0],xopt[1], color="black", s=12)
```

```python
# Minimizar f(x,y) = (a -x)**2 + b*(y - x**2)**2
def f(var):
    a,b = (1,100)
    x,y = var
    f = (a - x)**2 + b*(y - x**2)**2
    return f
```

```python
x0 = [-1.2,1]

xopt = opt.fmin(f,x0)
```

```python
print(xopt)
```

```python
xopt = opt.minimize(f,x0)
print(xopt)
```
