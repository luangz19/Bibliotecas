```python
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
# Minimizar a função
y = lambda x: 2*x**2 + 20*x - 22
```

```python
xmin = -20
xmax = 20

x_min = optimize.fminbound(y,xmin,xmax)
y_min = y(x_min)
print(x_min,y_min)
```

```python
res = optimize.minimize_scalar(y)
print(res)
```

```python
x = [x for x in range(-20,21)]
y = [y(x) for x in range(-20,21)]
```

```python
plt.style.use("dark_background")
plt.plot(x,y)
plt.scatter(x_min,y_min, c="r")
```

```python
# Maximizar a função y = -2*x**2 + 20*x - 22
y = lambda x: -2*x**2 + 20*x - 22
y_negat = lambda x: -(-2*x**2 + 20*x - 22)
```

```python
xmin = -20
xmax = 20
x_max = optimize.fminbound(y_negat,xmin,xmax)
y_max = -y_negat(x_max)
print(x_max,y_max)
```

```python
res = optimize.minimize_scalar(y_negat)
print(res)
```

```python
x = [x for x in range(-20,21)]
y = list(map(y,x))
```

```python
plt.plot(x,y)
plt.scatter(x_max,y_max,c="r")
```
