```python
import scipy.optimize as sc
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

Um fazendeiro tem 1 200 m de cerca e quer cercar um campo retangular que
está na margem de um rio reto. Ele não precisa de cerca ao longo do rio. 

Quais são as dimensões do campo que tem $\textbf{maior área}$?

Aréa do retângulo:
$A = x \cdot y$, onde $x$ é a altura e $y$ é a base

Perímetro: $1200 = 2x + y$
  
logo $y = 1200 - 2x$ 

$A = -2x^2 + 1200x$
  
  

```python
x = sp.Symbol('x')
y = sp.Function("y")(x)
y
```

```python
dAdx = y.diff(x)
dAdx
```

```python
pc = sp.solve(dAdx)
pc
```

```python
d2Adx2 = dAdx.diff(x)
d2Adx2
```

```python
plt.style.use("dark_background")
sp.plot(y,(x,0,600))
```

```python
# Máximizar uma função f(x) equivale a minimizar (-f(x))
A = lambda x: -(-2*x**2 + 1200*x)
```

```python
xmin = 0
xmax = 400
```

```python
xmax = sc.fminbound(A,xmin,xmax)
ymax = -A(xmax)
# Assim, o campo retangular deve 
# ter 300m de profundidade e 600m de extensão
print(xmax,ymax)
```

```python
x = [x for x in range(0,600)]
y = [-A(x) for x in range(0,600)]
```

```python
plt.plot(x,y)
plt.scatter(xmax,ymax, color="r")
```
