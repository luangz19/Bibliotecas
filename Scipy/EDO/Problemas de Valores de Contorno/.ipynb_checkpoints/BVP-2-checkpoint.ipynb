```python
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
```

```python
# y" + k**2 * y = 0
# y(0) = y(1) = 0
# y'(0) = k

# y[0] = y
# y[1] = y'
# y" = - k**2 * y

# Definindo o modelo
def Model(x,y,p):
    k = p[0]
    return np.vstack((y[1], -k**2*y[0]))
```

```python
# Definindo as condições contorno
def bc(ya,yb,p):
    k = p[0]
    return np.array([ya[0], yb[0], ya[1] - k])
```

```python
# encontrar a solução para k = 2 * pi, 
# para conseguir isso definimos valores de y 
# para seguir aproximadamente sin(2 * pi * x):

x = np.linspace(0,1,5)
y = np.zeros((2,x.size))
y[0,1] = 1
y[0,3] = -1
```

```python
sol = solve_bvp(Model, bc, x, y, p=[6])
```

```python
sol.p[0]
```

```python
x_plot = np.linspace(0,1,100)
y_plot = sol.sol(x_plot)[0]
plt.style.use("dark_background")
plt.plot(x_plot,y_plot)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
