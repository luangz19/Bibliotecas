```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp
```

```python
# y" + y = 0, y(0) = 0, y(pi) = 0
# y[0] = y
# y[1] = y'
# y" = -y

# Definindo o modelo
def Model(x,y):
    return np.vstack((y[1],-y[0]))
```

```python
# Definindo as condições de contorno
def bc(ya,yb):
    # y(0) = 0, y(pi) = 0
    return np.array([ya[0],yb[0]])
```

```python
x = np.linspace(0,np.pi,10)
y_a = np.zeros((2,x.size))
y_b = np.zeros((2,x.size))
y_b[0] = 2
```

```python
sol1 = solve_bvp(Model,bc,x,y_a)
sol2 = solve_bvp(Model,bc,x,y_b)
```

```python
x_p = np.linspace(0,6,100)
y_a = sol1.sol(x_p)[0]
y_b = sol2.sol(x_p)[0]
```

```python
plt.style.use("dark_background")
plt.plot(x_p,y_a)
plt.plot(x_p,y_b)
```
