```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp
```

```python
# y" + 2y = 0, y(0) = 1, y(pi) = 0
# y[0] = y
# y[1] = y'
# y" = -2y

# Definindo o modelo
def Model(x,y):
    return np.vstack((y[1],-2*y[0]))
```

```python
# Definindo as condições de contorno
def bc(ya,yb):
    return np.array([ya[0]-1,yb[0]])
```

```python
x = np.linspace(0,np.pi,10)
y = np.zeros((2,x.size))
```

```python
sol = solve_bvp(Model,bc,x,y)
```

```python
x_plot = np.linspace(0,np.pi,100)
y_plot = sol.sol(x_plot)[0]
plt.style.use("dark_background")
fig,ax = plt.subplots(2,1, figsize=(6,5))
ax[0].plot(x_plot,y_plot, label="Solução Númerica")
ax[0].legend()
ax[0].grid(linestyle=":", alpha=0.5)

f = np.cos(np.sqrt(2)*x_plot)-(np.cos(np.sqrt(2)*np.pi)/np.sin(np.sqrt(2)*np.pi)*np.sin(np.sqrt(2)*x_plot))
ax[1].plot(x_plot,f, label="Solução Analítica")
ax[1].legend()
ax[1].grid(linestyle=":", alpha=0.5)

plt.tight_layout()
```
