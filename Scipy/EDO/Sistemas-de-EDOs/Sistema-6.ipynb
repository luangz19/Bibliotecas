```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# dx1dt = x1 + 2x2
# dx2dt = -2x1 + x2
```

```python
A = np.array([[1,2],[-2,1]])
dXdt = lambda t,x: A @ x
```

```python
x0 = [1.0,0.5]
t = np.linspace(-10,10,1000)
```

```python
sol = solve_ivp(dXdt,[-10,10],x0,t_eval=t)
```

```python
plt.style.use("dark_background")
fig,ax = plt.subplots(2,1)
ax[0].plot(sol.t,sol.y[0])
ax[1].plot(sol.t,sol.y[1])
plt.tight_layout()
```

```python
sol2 = [solve_ivp(dXdt,[-10,10],y0,t_eval=t) for y0 in ([2.0,1.0],[1.0,0.5],[0.5,0.3])]
```

```python
for i in range(len(sol2)):
    plt.plot(sol2[i].y.T[:,0],sol2[i].y.T[:,1])
```

```python
# dx1dt = x1 + 2x2
# dx2dt = -2x1 + x2
x,y = np.linspace(-10,10,20),np.linspace(-10,10,20)
X1,X2 = np.meshgrid(x,y)
dx1 = X1 + 2*X2
dx2 = -2*X1 + X2

M = (dx1**2 + dx2**2)**(1/2)

DX1 = dx1/M
DX2 = dx2/M

plt.quiver(X1,X2,DX1,DX2,color="yellow")
```
