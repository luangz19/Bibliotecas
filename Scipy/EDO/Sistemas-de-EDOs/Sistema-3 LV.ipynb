```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def LotkaVolterra(t,h):
    a,b,k,r = 0.001, 0.00002, 0.08, 0.02
    C,L = h
    v1 = k*C - a*C*L
    v2 = -r*L + b*C*L
    return [v1,v2]
    
t = np.linspace(0,170,500)
sol2 = solve_ivp(LotkaVolterra, [0,1000], [1000,40], t_eval=t)
```

```python
plt.style.use('dark_background')
plt.plot(sol2.y.T[:,0],sol2.y.T[:,1])
plt.xlabel("C(t)")
plt.ylabel("L(t)")
plt.title("Retrado de faze C vs L")
plt.tight_layout()
```
