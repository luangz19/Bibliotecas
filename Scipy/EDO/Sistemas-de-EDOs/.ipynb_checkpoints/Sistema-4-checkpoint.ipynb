```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# Sistema Linear
# x1' = -2x1 + 3x2
# x2' = x1 - 4x2
```

```python
A = lambda t,S: np.dot(np.array([[-2,3],[1,-4]]),S)
```

```python
z0 = [1,2]
t = np.linspace(-10,10,100)
```

```python
sol = solve_ivp(A,[-10,10],z0,t_eval=t)
```

```python
plt.style.use("dark_background")
fig,ax = plt.subplots(2,1)
ax[0].plot(sol.t,sol.y[0])
ax[1].plot(sol.t,sol.y[1])
plt.tight_layout()
```

```python
x1 = lambda t: 3*C1*np.exp(-t) + C2*np.exp(-5*t)
x2 = lambda t: 3*C1*np.exp(-t) - C2*np.exp(-5*t)
```

```python
x,y = np.linspace(-10,10,20),np.linspace(-10,10,20)
X1,X2 = np.meshgrid(x,y)

dx1 = -2*X1 + 3*X2
dx2 = X1 - 4*X2

M = np.sqrt(dx1**2 + dx2**2)
dx1 = dx1/M
dx2 = dx2/M

plt.figure(figsize=(8,6))
plt.quiver(X1,X2,dx1,dx2, color="yellow")
plt.grid(linestyle=":", alpha=0.5)
for i in [(1,0),(0,1),(1,1),(-2,2),(-2,3)]:
    C1,C2 = i
    h = np.linspace(0,10,100)
    yp = x1(h)
    yp2 = x2(h)
    plt.plot(h,yp)
    plt.plot(h,yp2)
```
