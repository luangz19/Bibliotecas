```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def Y(t,z):
    x, y = z
    dxdt = t*x + y
    dydt = -t**2*x - t*y
    return [dxdt, dydt]

sol = solve_ivp(Y, [0,25], [0,4], dense_output=True) 
#[0,10]: intervalo de tempo (de t=0 a t=10)
#[0,4]: condições iniciais (x=0, y=4 em t=0)

t = np.linspace(0,25,100)
z = sol.sol(t)

plt.style.use('dark_background')
#plt.plot(t,z.T)
plt.plot(t,z[0])
ax1 = plt.axes([.13,.25,.77,.5], facecolor='none')
ax1.plot(t,z[1], color='red')
ax1.yaxis.tick_right()                   # Coloca os valores do eixo y para outro lado
ax1.set_xticks([])                       # Remove os valores do eixo x
ax1.spines['bottom'].set_visible(False)  # Remove borda inferior
ax1.spines['left'].set_visible(False)    # Remove borda esquerda
ax1.spines['top'].set_visible(False)     # Remove borda superior
```

```python
t = np.linspace(-34,45,1000)
sol2 = solve_ivp(F,[-34,45],[1,4], t_eval=t) 
plt.plot(sol2.y.T[:,0],sol2.y.T[:,1])
```

```python
t,x,y = 2.0, np.linspace(-1,1,20), np.linspace(-1,1,20)

X,Y = np.meshgrid(x,y)

dx = t*X + Y
dy = -t**2*X - t*Y

M = np.sqrt(dx**2 + dy**2)

dx /= M
dy /= M

plt.quiver(X,Y,dx,dy, color="yellow")
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Campo de vetores no espaço de fase (t = {t})")
plt.grid(True)
plt.show()
```
