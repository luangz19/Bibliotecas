```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def Y(t,z):
    x, y = z
    dxdt = t*x + y
    dydt = -t**2*x - t*y
    return [dxdt, dydt]

sol = solve_ivp(Y, [0,25], [0,4], dense_output=True) 
#[0,10]: intervalo de tempo (de t=0 a t=10)
#[0,4]: condições iniciais (x=0, y=4 em t=0)

t = np.linspace(0,25,100)
z = sol.sol(t)

plt.style.use('dark_background')
#plt.plot(t,z.T)
plt.plot(t,z[0])
ax1 = plt.axes([.13,.25,.77,.5], facecolor='none')
ax1.plot(t,z[1], color='red')
ax1.yaxis.tick_right()                   # Coloca os valores do eixo y para outro lado
ax1.set_xticks([])                       # Remove os valores do eixo x
ax1.spines['bottom'].set_visible(False)  # Remove borda inferior
ax1.spines['left'].set_visible(False)    # Remove borda esquerda
ax1.spines['top'].set_visible(False)     # Remove borda superior
```

```python
def LotkaVolterra(t,h):
    a,b,k,r = 0.001, 0.00002, 0.08, 0.02
    C,L = h
    v1 = k*C - a*C*L
    v2 = -r*L + b*C*L
    return [v1,v2]

sol2 = solve_ivp(LotkaVolterra, [0,1000], [1000,40],dense_output=True)
t2 = np.linspace(0,500,100)
h = sol2.sol(t2)
plt.plot(t2,h[0])

ax2 = plt.axes([.13,.25,.77,.5], facecolor='none')
ax2.plot(t2,sol2.sol(t2)[1],color='red')
ax2.yaxis.tick_right()                   # Coloca os valores do eixo y para outro lado
ax2.set_xticks([])                       # Remove os valores do eixo x
ax2.spines['bottom'].set_visible(False)  # Remove borda inferior
ax2.spines['left'].set_visible(False)    # Remove borda esquerda
ax2.spines['top'].set_visible(False)     # Remove borda superior
```
