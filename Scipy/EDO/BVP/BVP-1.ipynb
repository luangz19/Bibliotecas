```python
from scipy.integrate import solve_bvp
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib qt
```

```python
# y" + k*exp(y) = 0, onde k=1, BVP y(0) = 0 e y(1) = 0
# y" = -k*exp(y)
# y0 = y, y0' = y' = y1
# y1 = y', y1' = y" = -k*exp(y) = -k*exp(y0)
# [y1,-exp(y0)]

# y[0] = y    
# y[1] = y'  
# y" = -exp(y)  

# Definindo o modelo
def Model(x,y):
    # y[0] = y, y[1] = y'      
    # y" = -exp(y)   y[0]' = y[1]', y1' = -exp(y[0])
    return np.vstack((y[1], -np.exp(y[0])))

# Condições de Contorno
def bc(ya,yb):
    # y(0) = 0 e y(1) = 0
    return np.array([ya[0],yb[0]])
```

```python
x = np.linspace(0,1,5)
y_a = np.zeros((2,x.size))
y_b = np.zeros((2,x.size))
y_b[0] = 3
```

```python
res_a = solve_bvp(Model, bc, x, y_a)
res_b = solve_bvp(Model, bc, x, y_b)
```

```python
x_plot = np.linspace(0,1,100)
y_plot_a = res_a.sol(x_plot)[0]
y_plot_b = res_b.sol(x_plot)[0]
```

```python
plt.style.use("dark_background")
plt.plot(x_plot, y_plot_a)
plt.plot(x_plot,y_plot_b)
```
