```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import sympy as sp
```

```python
def dxdt(t,x):
    dxdt = 3*x**(2/3)
    return dxdt
```

```python
y0 = 0
t_span = np.linspace(-2,2,100)
```

```python
plt.style.use("dark_background")
for yi in range(0,50,5):
    yi = yi
    sol = solve_ivp(dxdt,[-2,2],[yi], t_eval=t_span)
    plt.plot(sol.t,sol.y[0], color="orange", alpha=0.6, linestyle="-")

plt.grid(linestyle=":",alpha=0.5)
plt.tight_layout()
```

```python
x = np.linspace(0,3,20)
t = np.linspace(-10,10,20)
X,T = np.meshgrid(x,t)
DT = 1
DX = 3*X**(2/3)
M = np.sqrt(DX**2 + DT**2) + 1e-9
DT = DT/M
DX = DX/M
```

```python
plt.figure(figsize=(4,4))
for yi in range(0,12,2):
    yi = yi/10
    t_span = np.linspace(0,1.5)
    sol = solve_ivp(dxdt,[0,1.5],[yi], t_eval=t_span)
    plt.plot(sol.t,sol.y[0], color="white", alpha=0.6, linestyle=":")
    
plt.quiver(X,T,DT,DX, color="yellow")
plt.ylim(-1,10)
plt.tight_layout()
```
