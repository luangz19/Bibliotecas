```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

### Equação Logística
- $r$ é taxa de  crescimento
- $K$ é capacidade de suporte
- 
$ \dfrac{dP}{dt} = rP \displaystyle \left(1 - \frac{P}{K} \right) $

```python
F = lambda t,P: 0.08*P*(1 - P/1000)
t = np.linspace(0,100,100)
solF = solve_ivp(F,[0,100],[2500], t_eval=t) #y0 = 2500
solF2 = solve_ivp(F,[0,100],[500], t_eval=t) #y0 = 500
plt.style.use("dark_background")
plt.plot(solF.t,solF.y[0])
plt.plot(solF2.t,solF2.y[0])
plt.tight_layout()
```

### Equação Gompertz
- $r$ é taxa de  crescimento intesítrica
- $K$ é capacidade de suporte
- $ \dfrac{dP}{dt} = rP \displaystyle \ln {\left(\frac{K}{P} \right)} $

```python
G = lambda t,P: 0.08*P*np.log(1000/P)
t = np.linspace(0,100,100)
solG = solve_ivp(G,[0,100],[2500], t_eval=t) #y0 = 2500
solG2 = solve_ivp(G,[0,100],[500], t_eval=t) #y0 = 500
plt.plot(solG.t,solG.y[0])
plt.plot(solG2.t,solG2.y[0])
plt.tight_layout()
```

```python
fig,ax = plt.subplots(1,2)
ax[0].plot(solF.t,solF.y[0], label = "Logística")
ax[0].plot(solG.t,solG.y[0], label = "Gompertz")
ax[0].legend()

ax[1].plot(solF2.t,solF2.y[0], label = "Logística")
ax[1].plot(solG2.t,solG2.y[0], label = "Gompertz")
ax[1].legend()


plt.tight_layout()
```
