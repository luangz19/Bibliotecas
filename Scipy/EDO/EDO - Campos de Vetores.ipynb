```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def dxdt(t,x):
    return 2*t*x
```

```python
t,x = np.linspace(0,2,20),np.linspace(-1,1,20)
T,X = np.meshgrid(t,x)
DT = 1
DX = dxdt(T,X)

magnitude = np.sqrt(DT**2 + DX**2)
DT /= magnitude
DX /= magnitude
```

```python
plt.style.use("dark_background")
for i in range(0,10):
    t_span = np.linspace(0,1.7,100)
    y0 = (-1)**i * (i/100)
    sol = solve_ivp(dxdt,[0,2],[y0], t_eval=t_span)
    plt.plot(sol.t,sol.y[0])
plt.quiver(T,X,DT,DX,color="yellow")
plt.xlim(0,1.75)
plt.ylim(-1,1)
plt.tight_layout()
```

```python
def dPdt(t,P):
    K = 1000
    r = 0.08
    return r*P*(1 - P/K)
```

```python
t,p = np.linspace(0,40,20), np.linspace(0,1400,20)
T,P = np.meshgrid(t,p)

DT = 1
DP = dPdt(t,P)


magnitude = np.sqrt(DT**2 + DP**2)
DT /= magnitude
DP /= magnitude

for i in range(0,16,2):
    t_span = np.linspace(0,40,100)
    y0 = i*100
    sol = solve_ivp(dPdt,[0,40],[y0], t_eval=t_span)
    plt.plot(sol.t,sol.y[0])
    
plt.quiver(T,P,DT,DP,color="yellow")
plt.xlim(0,40)
plt.ylim(0,1400)
plt.tight_layout()
```
