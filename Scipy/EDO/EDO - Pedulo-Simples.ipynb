```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def Pend(t,z):
    g = 10
    l = 1.25
    O,dOdt = z
    dOdt2 = -(g/l)*np.sin(O)
    return [dOdt,dOdt2]
```

```python
O0,dOdt0 = np.pi/4,0
z0 = [O0,dOdt0]
t = np.linspace(0,10,1000)
```

```python
sol = solve_ivp(Pend,[0,10],z0,t_eval=t)
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(8, 4))
plt.plot(sol.t, sol.y[0], label='ngulo (rad)')
plt.plot(sol.t, sol.y[1], label='Velocidade angular (rad/s)')
plt.xlabel('Tempo (s)')
plt.ylabel('Valor')
plt.title('Pndulo simples')
plt.legend()
plt.grid(True)
plt.show()
```
