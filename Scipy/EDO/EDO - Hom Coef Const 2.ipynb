```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

#### EDO com Coeficientes constantes
$y'' + 4y = 0$

$\begin{cases}
y(0) = 10
\\ \\
y'(0) = 0
\end{cases}$


```python
def Model(t,z):
    y, dydt = z
    dydt2 = -4*y
    return [dydt,dydt2]
```

```python
# Condição Inicial
z0 = [10,0]
t = np.linspace(0,10,100)
sol = solve_ivp(Model,[0,10],z0, t_eval=t)
```

```python
plt.style.use("dark_background")
plt.plot(sol.t,sol.y[0])
```

$y'' + \sqrt{7}y' + 4y = 0$

$\begin{cases}
y(0) = 10
\\ \\
y'(0) = 0
\end{cases}$

```python
def Model2(t,z):
    y,dydt = z
    dydt2 = -np.sqrt(7)*dydt - 4*y
    return [dydt,dydt2]

z = [10,0]
t = np.linspace(0,15,100)
sol2 = solve_ivp(Model2,[0,15], z0,t_eval=t)

plt.plot(sol2.t,sol2.y[0])
```

```python
import sympy as sp
x = sp.Symbol('x')
y = sp.Function('y')('x')
eq_diff = sp.Eq(y.diff(x,2) + 4*y,0)
```

```python
ics = {y.subs(x,0):10,y.diff(x).subs(x,0):0}
sol = sp.dsolve(eq_diff,y, ics=ics)
sol.rhs
```

```python
sp.plot(sol.rhs)
```

```python
eq_diff2 = sp.Eq(y.diff(x,2) + sp.sqrt(7)*y.diff(x) + 4*y,0)
eq_diff2
```

```python
ics = {y.subs(x,0):10,y.diff(x).subs(x,0):0}
sol2 = sp.dsolve(eq_diff2,y, ics=ics)
sol2.rhs
```

```python
sp.plot(sol2.rhs,(x,0,10))
```
