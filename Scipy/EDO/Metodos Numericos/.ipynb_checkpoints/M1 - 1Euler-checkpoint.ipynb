### Método de Euler 

É o método mais simples para a aproximação da solução de EDO's.
Considere o seguinte problema de valor inicial de primeira ordem:

$\begin{align}
\begin{cases}
y' = \dfrac{dy}{dx} = f(t,y) \\ \\
y(t_0) = y_0 
\end{cases}
\end{align}$

A solução deste PVI passa obrigatoriamente pelo ponto $(t_0 , y_0 )$. O objetivo do método de
Euler é estimar o valor da solução $y(t)$ em um ponto com uma distância $\Delta t$ do ponto inicial
$t0$ . Este ponto passa a ser chamado de $t_1 = t_0 + ∆t$, de modo que o do método de Euler é
determina o valor da solução em $t_1$ , $y_1 = y(t_1).$

A partir deste valor, pode-se estimar o valor
da função em um ponto $t_2$ distante $\Delta t$ do ponto $t_1$ e continuar com este procedimento até
se obter uma aproximação para a solução por quantos pontos forem necessários.

$y_{n+1} = y_n + \Delta t f(t_n,y_n)$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## dydt =  -2t - y  com  Dt = 0.1, y(0) = -1

Dt = 0.1   # Tamanho do passo
N = 21     # Número de pontos
y = np.zeros(N) # array com N entradas zeros
t = np.linspace(0,Dt*(N-1),N) # array com N entradas igualmente espaçados com tamnaho de 0.1
y[0] = -1  # condição inicial

# Método de Euler
for i in range(N - 1):
    y[i + 1] = y[i] + Dt*(-2*t[i] - y[i])

# Solução Analítica
y_anal = -2*t + 2 -3*np.exp(-t)

plt.style.use("dark_background")
plt.plot(t,y, label="Sol. Num.")
plt.plot(t,y_anal, label="Sol. Anal.")
plt.scatter(t[10], y[10])
plt.scatter(t[10], y_anal[10])
plt.legend()
```

```python
Erro_Abs = pd.DataFrame({"Solu. Numérica":y, 
                     "Solu. Analítica":y_anal, 
                     "Diferença Abs": abs(y - y_anal)})
Erro_Abs
```
