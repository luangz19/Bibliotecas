```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Metodo de RK4
# dydt = −y + 2t, y(0) = 2

N = 1000
t = np.linspace(0, 1, N+1)
h = 1.0/N
h2 = h / 2

# inicialização
#t = np.zeros(N+1)
y = np.zeros(N+1)
y[0] = 2

for i in range(N):
    k1 = -y[i] + 2 * t[i]
    k2 = -(y[i] + h2*k1) + 2 * (t[i] + h2)
    k3 = -(y[i] + h2*k2) + 2 * (t[i] + h2)
    k4 = -(y[i] + h*k3) + 2 * (t[i] + h)
    y[i+1] = y[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
```

```python
plt.style.use("dark_background")
plt.plot(t,y)
```

```python
dydt = lambda t,y: -y + 2*t #y(0) = 2

sol = solve_ivp(dydt, [0,1], [2], t_eval=t, method="RK45")

plt.plot(sol.t,sol.y[0])
```
