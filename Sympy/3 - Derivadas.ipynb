### Derivdas de funções em $\mathbb{R}$
$ \dfrac{df}{dx} = f'(x) = \displaystyle \lim_{x \to h} \frac{f(x + h) -f(x)}{h}$

```python
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x,a,b,c = sp.symbols('x a b c')
```

```python
f = 3*x + 4
f
```

```python
# derivda de f
dfdx = sp.diff(f,x)
dfdx
```

```python
plt.style.use("dark_background")
sp.plot(f,dfdx,(x,-5,5),label=(f"f(x) = {f}",f"f'(x) = {dfdx}"),legend=True)
```
