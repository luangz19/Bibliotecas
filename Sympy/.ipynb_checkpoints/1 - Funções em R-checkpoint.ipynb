```python
#Função Quadrática
f = a*x**2 + b*x + c
f
```

```python
# Função Exponencial
exp = sp.exp(x)
exp
```

```python
sp.plot(exp, (x,-5,5))
```

```python
# Função Logarítima
log = sp.log(x)
log
```

```python
sp.plot(log,(x,-5,5))
```

```python
# Função Seno
sen = sp.sin(x)
sen
```

```python
sp.plot(sen, (x,-5,5))
```

```python
# Função Cosseno
cos = sp.cos(x)
cos
```

```python
sp.plot(cos, (x,-5,5))
```

```python
tan = sp.tan(x)
tan
```

```python
sp.plot(tan,(x,-2*sp.pi,2*sp.pi))
```

```python
raiz = sp.sqrt(x)
raiz
```

```python
sp.plot(raiz,(x,0,5))
```
