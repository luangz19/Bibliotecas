```python
import matplotlib.pyplot as plt
import sympy as sp
from sympy.plotting import plot3d,plot_contour
```

```python
x = sp.Symbol('x')
y = sp.Function("y")(x)
```

```python
dydx = sp.Eq(y.diff(x), x/y)
```

```python
sol = sp.dsolve(dydx)
```

```python
sol[0]
```

```python
sol[1]
```

```python
PVI = {y.subs(x,0):1}
sol1 = sp.dsolve(dydx, ics = PVI)
```

```python
PVI = {y.subs(x,0):-1}
sol2 = sp.dsolve(dydx, ics = PVI)
```

```python
sol1
```

```python
sol2
```

```python
# C1,_ = tuple(sol[0].free_symbols)
```

```python
y1 = sol1.rhs
y2 = sol2.rhs
```

```python
plt.style.use("dark_background")
sp.plot(y1,y2, (x,-1,1))
```

```python
x,y = sp.symbols("x y")
```
