```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def dydx(x,y):
    return -2*(x/y)
```

```python
# y0 = 1
# z0 = -1
t = np.linspace(-1,1,100)
sol1 = [solve_ivp(dydx,[-1,1],[y0],t_eval=t) for y0 in [0.5,1.0,1.5]]
sol2 = [solve_ivp(dydx,[-1,1],[z0],t_eval=t) for z0 in [-0.5,-1.0,-1.5]]
```

```python
plt.style.use("dark_background")
for i in range(len(sol1)):
    plt.plot(sol1[i].t,sol1[i].y[0])
    plt.plot(sol2[i].t,sol2[i].y[0])
plt.grid(alpha=0.2)
plt.tight_layout()
```

```python
# Solução Analítica
# y² = -2x² + C
# f(x,y) = y² + 2x²
```

```python
x,y = np.linspace(-2,2,100),np.linspace(-2,2,100)
X,Y = np.meshgrid(x,y)
Z = 2*X**2 + Y**2
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(121,projection="3d")
ax2 = fig.add_subplot(122)
ax1.plot_surface(X,Y,Z)
ax2.contour(X,Y,Z)
ax2.grid(linestyle=":", alpha=0.5)

plt.tight_layout()
```
