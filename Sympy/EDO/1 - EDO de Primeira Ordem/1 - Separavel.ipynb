### Equações separáveis

$\dfrac{dy}{dx} = f(x,y)$  onde ($x$ é a variável independente é variável $y$ dependente)

- $h(y)dy = g(x)dx$
- $f(x,y) = g(x)\frac{1}{h(y)}$

$\dfrac{dy}{dx} = g(x)p(y)$ onde $p(y) = \frac{1}{h(y)}$

```python
import sympy as sp
from matplotlib import style
```

```python
# Criando a variável simbólica
x = sp.Symbol('x')

# Criando a função y(x)
y = sp.Function("y")(x)
y
```

```python
# Defininfo a EDO
dydx = sp.Eq(y.diff(x),x*y)
dydx
```

```python
# Resolvendo a EDO
y_x = sp.dsolve(dydx)
y_x
```

```python
# Criando um PVI
PVI = {y.subs(x,0):1} # y(0) = 1
PVI
```

```python
# Resolvendo a EDO com o PVI
y_x = sp.dsolve(dydx, ics=PVI)
y_x
```

```python
# Pegando o lado direito da equação
y = y_x.rhs
y
```

```python
style.use("dark_background")
sp.plot(y)
```

```python
# Criando a variável simbólica
t = sp.Symbol('t')

# Criando a função y(x)
y = sp.Function("y")(t)
y
```

```python
# Defininfo a EDO
dydt = sp.Eq(y.diff(t),y*(1-y))
dydt
```

```python
# Resolvendo a EDO
y_x = sp.dsolve(dydx)
y_x
```

```python
# Criando o PVI
PVI = {y.subs(t,0):2}
PVI
```

```python
# Resolvendo a EDO com o PVI
y_t = sp.dsolve(dydt, ics=PVI)
y_t
```

```python
y = y_t.rhs
y
```

```python
limy = sp.limit(y,t,-sp.oo) #lim y quando t --> -oo
limy
```

```python
sp.plot(y,(t,-0.5,0))
```

```python
limy = sp.limit(y,t,sp.oo) #lim y quando t --> +oo
limy
```

```python
sp.plot(y,(t,0.5,0))
```

```python
# Criando a variável simbólica
t = sp.Symbol('t') # tempo

# Criando a função y(x)
y = sp.Function("y")(t) # População

# Modelo logístico
r = 0.06 # taxa de crescimento
K = 1000 # capacidade de suporte
dydt = sp.Eq(y.diff(t), r*y*(1 - y/K))
dydt
```

```python
# Resolvendo a EDO
y_t = sp.dsolve(dydt)
y_t
```

```python
# Criando o dois PVI
PVI_1 = {y.subs(t,0):40} # onde y(0) = 40 significa que a população é de 40 indivíduos
PVI_2 = {y.subs(t,0):2500} # onde y(0) = 2500 significa que a população é de 2500 indivíduos
```

```python
# Resolvendo a EDO com o PVI_1
y_t1 = sp.dsolve(dydt, ics=PVI_1)
y_t1
```

```python
# Resolvendo a EDO com o PVI_2
y_t2 = sp.dsolve(dydt, ics=PVI_2)
y_t2
```

```python
y1 = y_t1.rhs
y1
```

```python
y2 = y_t2.rhs
y2
```

```python
# se t = anos então a população 
# que começou no instante y(0) = 40 
# terá aproximadamente mais de 400 indivíduos
# quando t = 50 anos
sp.plot(y1, (t,0,50)) 
```

```python
# quando for t aproximadamente t = 125 anos
# os recursos começam a ficar limitados
sp.plot(y1, (t,0,125)) 
```

```python
# se que começou no instante y(0) = 2500
# terá aproximadamente mais de 1200 indivíduos
# quando t = 50 anos, isto significa que os recursos
# ficaram limitados e a população entra em declínio
sp.plot(y2, (t,0,50)) 
```

```python
sp.plot((y1,(t,0,50)),(y2,(t,0,50)))
```

```python
# aumentando t para as duas funções
sp.plot((y1,(t,0,125)),(y2,(t,0,90)))
```
