### Integrais de funções em $\mathbb{R}$
$ \displaystyle \int_a^b f(x)\,dx = F(x) + C$

```python
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x,a,b,c = sp.symbols('x a b c')
```

```python
f = 3*x + 4
f
```

```python
# integral indefinida de f
Sf = sp.integrate(f,x) 
Sf
```

```python
plt.style.use("dark_background")
sp.plot(f,Sf,(x,-5,5),label=(f"f(x) = {f}",f"F(x) = {Sf}"),legend=True)
```
